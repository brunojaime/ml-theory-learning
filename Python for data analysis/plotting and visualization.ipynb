## Adjusting the spacing around subplot
By default matplolib leaves an amount of padding around subplots and in spacing between subplots. The plot will dinamically adjust itself. This spacing can be changed using subplots_adjust

```python
# wspace and hspace control the percent of the figure height and widtth to use as spacing between plots
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500),bins=50,color="black",alpha=0.5)
fig.subplots_adjust(wspace=0.2,hspace=0.2)
```

# Ticks, labels and legends
Most of plot decorations can be accessed through methods on matplotlib axes objects. 
- xlim -> plot range
- xticks -> tick locations
- xticklabels -> tick labels
Two wways to call them
- No arguments -> returns the current parameter value
- With parameter -> Sets the parameter
All such methods act on the most recently created AxesSubPlot




```python
#Setting title, axis labels, ticks and tick labels
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
#tick= ax.set_xticks(np.linspace(250,1000,4))
tick= ax.set_xticks(np.arange(0,1250,250))
tick_y = ax.set_yticks(np.arange(-20,50,5))
labels= ax.set_xticklabels(['one','two','three','four','five'],rotation=10,fontsize=10)
set_title= ax.set_xlabel("Stages")
title_y = ax.set_ylabel("Prices")
ax.set_title("Grfico con los labels correspondientes")
```

```python
fig, ax= plt.subplots()
ax.plot(np.sqrt(np.arange(0,100,2)))
ax.set(title="Grfico de races cuadradas",xlabel="Stages",ylabel="Money")
```

### Density plot
Computed as an estimate of a continuous probability that might have generate the observed data. Usual procedure is to approximate the distribution as a mixture of kernels.
Also known as Kernel Density Estimate (KDE) plots.


```python
fig,axes = plt.subplots()
values =np.random.standard_normal(1000)
axes.hist(values,bins=30)
```

```python
df = pd.Series(values)
df.plot.density()
```

## Scatter or point plots

# Facet Grids and Categorical Data
Facet grids-> It allows us to visualize data with many categorical variables. Two dimensional layout of plots where the data is split across the plot on each axis based on the distinct values of a certain variable. seaborn -> catplot function

```python
import random
dias=['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo']
tipo = ['Almuerzo','Cena']
size= 90

cantidad = np.random.randint(20,80,size=size)
days= np.random.choice(dias,size=size,replace=True)
hue= np.random.choice(tipo,size=size,replace=True)
fuma = np.random.choice([True,False],size=size,replace=True)
df = pd.DataFrame({'dia':days,'hora':hue,'fuma':fuma,"cantidad":cantidad})
```

```python
df = pd.DataFrame({'dia':days,'hora':hue,'fuma':fuma,"cantidad":cantidad})
```

```python
df.head()
```

```python
sns.catplot(x='dia',y='cantidad',hue='hora',col='fuma',kind='cbar',data=df)
```

```python
sns.catplot(x='dia',y='cantidad',row="hora",col="fuma",kind="bar",data=df)
```

```python
# Other plot types
#Box plots (median,quartiles and outliers)
sns.catplot(x="cantidad",y="dia",kind="box",data=df)
```
