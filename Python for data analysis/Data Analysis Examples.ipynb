```python
indexer = agg_counts.sum(axis="columns").argsort() #Gets an index array that can be used to sort in ascending order
indexer
```

```python
count_subset=agg_counts.take(indexer[-10:]) #We take the last 10 rows
count_subset
```

```python
#Pandas has a convienence method called nlargest 
agg_counts.sum(axis="columns").nlargest(10)
```

```python
count_subset = count_subset.stack()
count_subset.name ="Total"
count_subset
```

```python
count_subset = count_subset.reset_index() #This way we have better compatibility with seaborn
```

```python
count_subset.head()
```

```python
sns.barplot(y='tz', x= "Total",hue='os',data=count_subset)
```

```python
# Because it's difficult to see the relative percentage of the smallest groups, we can normalize the groups to sum 1
def norm_total(group):
    group['normed_total'] = group['Total']/group['Total'].sum()
    return group
results = count_subset.groupby('tz').apply(norm_total)
```

```python
sns.barplot(y='tz',hue="os",x="normed_total",data=results )
```

### MovieLnes 1M Dataset


```python
unames = ['user_id','gender','age','occupation',"zip"]
users = pd.read_table("datasets/movie_lens/users.dat",sep='::',names=unames,header=None,engine='python')

rnames = ['user_id',"movie_id","rating","timestamp"]
ratings = pd.read_csv("datasets/movie_lens/ratings.dat",sep="::",header=None,engine='python',names=rnames)

mnames = ['movie_id','title','genres']
movies = pd.read_csv("datasets/movie_lens/movies.dat",sep="::",header=None,engine='python',names=mnames)
```

```python
users.head()
users.info()
```

```python
ratings.head()
ratings.info()
```

```python
# We want to compute mean ratings for a particular movie by gender and age
# First lets merge the info
# We can merge user and ratings by user_id
#Then merge the result with movie by movie_id

data = pd.merge(pd.merge(users,ratings),movies,left_on="movie_id",right_on="movie_id",how="outer")
data
```

### Us Baby Names (1180-2010)
Frequency of baby names from 1880 until 2000

```python
dataset = pd.read_csv('datasets/baby_names/NationalNames.csv')
dataset.head(10)
```

```python
dataset = dataset.rename(columns={"Gender":"sex","Count":'births'})
dataset = dataset.rename(columns={"Name":"Name".lower(),"Year":'year'})
dataset
```

```python
total_birth =dataset.pivot_table('births',index='year',columns='sex',aggfunc='sum')
total_birth.tail(10)
```

```python
total_birth.plot(title="Total births by sex and year")
```

We'll insert a column prop with the fraction of babies given each name relative to the total number of births


```python
def prop_name_relative(group):
    group['prop'] = group['births'] / group['births'].sum()
    return group
```

```python
dataset = dataset.groupby(['year','sex']).apply(prop_name_relative)
#dataset = pd.DataFrame(dataset,index=['Id'])
dataset = dataset.reset_index(drop=True)
dataset
```

```python
dataset.groupby(['year','sex'])['prop'].sum()
```

Let's extract a subset of the data. Top 1000 names for each sex/year combination.

```python
def _1000Name(group):
    return group.sort_values('births',ascending=False)[:1000]
    
```

```python
data_1000_names = dataset.groupby(['year','sex']).apply(_1000Name,include_groups=True)
data_1000_names
```

```python
data_1000_names = data_1000_names.reset_index(drop=True)
data_1000_names
```

Number of distinct names, in order from highest to lowest in the 50% births

```python
df= boys[boys["year"]==2010]
df
```

```python
prop_cumsum =df['prop'].sort_values(ascending = False).cumsum()
prop_cumsum[:10]
```

```python
# searchsorted -> Returns the position in the cumulative sum at which 0.5 (parameter given)would need to be inserted to keep it in sorted order
prop_cumsum.searchsorted(0.5)
```

This means that the value prop with 0.5 can be inserted in the row 117 (counting the 0 index). But this is only for 2010. Let's see for 1950


```python
df = boys[boys['year']==1950]
prop_cumsum =df['prop'].sort_values(ascending=False).cumsum()
prop_cumsum.searchsorted(0.5) + 1
```

```python
# Lets make a function to see ir for every year
def prop_cum_sum(group,q=0.5):
    prop_cumsum_actual =group['prop'].sort_values(ascending=False).cumsum()
    return prop_cumsum_actual.searchsorted(q)+1
```

```python
diversity =data_1000_names.groupby(['year','sex']).apply(prop_cum_sum)
diversity = diversity.unstack()
diversity
```

```python
diversity.plot()
```

### Boys names that became girls names and viceversa
