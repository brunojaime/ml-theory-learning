Is hard to produce good visualizations.
There are two primary uses for data visualization:
- Explore data
- Communicate data

```python
import matplotlib.pyplot as plt
import numpy as np
```

In it's simple way, pyplot mantains an internal state in which you build up a visualization step by step.

```python
years = np.arange(7) * 10 + 1950
gdp = [300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]
```

```python
plt.plot(years,gdp,color="green",marker='o',linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("Years")
plt.show()
```

### Bar charts
Is good for showing quantity varies among some set of discrete items

```python
movies =["Esperando la carrosa","Blade II","Nueve Reinas","El secreto de sus ojos"]
num_oscars = [5,2,3,7]
```

```python
fig,ax = plt.subplots(figsize=(10,3))
ax.bar(movies,num_oscars)
ax.set_title("Oscars por película")
ax.set_ylabel("Cantidad")
ax.set_xlabel("Película")
plt.show()
```

Bar chart can also be a good choice for plotting histograms of bucketed numeric values, in order to visually explore how the values are distributed

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
histogram = Counter(min(grade // 10 * 10,90) for grade in grades)
histogram # Nos va a mostrar cuantos elementos hay de cada tipo
```

```python
plt.bar(
    [x + 5 for x in histogram.keys()], # Shift bars right by 5
    histogram.values(), # bar heights
    10, # Give each bar a width
    edgecolor=(0.4,1,0.7))
plt.axis([-5,105,0,5]) # axis x from -5 to 105 and y from 0 to 5

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("Number of students")
plt.title("Distribution of grades")
plt.show()
```

```python
mentions = [1000,1350]
years = [2020, 2024]

plt.bar([year + 0.8 /2 for year in years],mentions,0.8)
plt.xticks(years) #Solo muestro los años que corresponden
plt.ticklabel_format(useOffset=False)
plt.axis([2019,2025,600,1500])
```

### Line Charts
These are good for showing trends

```python
variance = 2 ** np.arange(9) 
bias_squared = variance[::-1]
total_error = [x + y for x,y in zip(variance,bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.plot(xs,variance,'g-',label="variance")
plt.plot(xs,bias_squared, 'rx',label="Bias-squared")
plt.plot(xs,total_error,'b:',label="total_error")
plt.legend(loc='best')
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The bias variance trade off")
plt.show()
```

### Scatterplots
is the right choice for visualizing the relationship between two paired sets of data.

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a','b','c','d','e','f','g','h','i']
```

```python
plt.scatter(friends,minutes)
for label,friend_count,minute_count in zip(labels,friends,minutes):
    plt.annotate(label,
                 xy=(friend_count,minute_count),
                 xytext=(5,-5),
                 textcoords="offset points")
                 
plt.title("Number of friends vs number of minutes on the website")
plt.xlabel("Number of friends")
plt.ylabel("Number of minutes")
```
